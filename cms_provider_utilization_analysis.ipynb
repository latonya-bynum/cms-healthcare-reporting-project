```python
# =========================================
# CMS Provider Utilization Analysis
# Author: LaTonya Bynum, MPH, MCHES
# =========================================

import os
import pandas as pd

# -------------------------------
# 1. Set File Path
# -------------------------------

file_path = r"C:\Healthcare_Reporting_Portfolio\01_raw_data\claims\provider.csv"

print("File path:", file_path)


# -------------------------------
# 2. Load Data
# -------------------------------
df = pd.read_csv(file_path, low_memory=False)

# -------------------------------
# 3. Basic Data Checks
# -------------------------------
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())

df.head()

# -------------------------------
# 4. Top Providers by Services
# -------------------------------
top_providers = (
    df.groupby('Rndrng_NPI')['Tot_Srvcs']
      .sum()
      .reset_index()
      .sort_values(by='Tot_Srvcs', ascending=False)
      .head(10)
)

print("\nTop Providers:")
print(top_providers)

# -------------------------------
# 5. Save Output
# -------------------------------
output_folder = r"C:\Healthcare_Reporting_Portfolio\04_outputs"

os.makedirs(output_folder, exist_ok=True)

output_file = os.path.join(output_folder, 'top_providers.csv')
top_providers.to_csv(output_file, index=False)

print("\nSaved to:", output_file)
```
